### 5. Distribution of Attacks by Year
This visualization shows how the number of recorded terrorist incidents has changed over time.

```python
import matplotlib.pyplot as plt
import seaborn as sns

if df is not None:
    plt.figure(figsize=(15, 8))
    sns.set_style('whitegrid')
    # Fixed the FutureWarning by assigning x to hue and setting legend=False
    sns.countplot(x='iyear', data=df, hue='iyear', palette='viridis', legend=False)
    plt.xticks(rotation=90)
    plt.title('Distribution of Terrorist Attacks by Year (1970-2017)')
    plt.xlabel('Year')
    plt.ylabel('Number of Attacks')
    plt.show()
```

## Data Cleaning and Feature Selection
To make the analysis more manageable, we will extract only the essential columns and handle basic missing values for the visualizations.

```python
if df is not None:
    # Selecting core columns for analysis
    cols_to_keep = [
        'iyear', 'imonth', 'iday', 'country_txt', 'region_txt', 'provstate',
        'city', 'latitude', 'longitude', 'success', 'suicide', 'attacktype1_txt',
        'targtype1_txt', 'targsubtype1_txt', 'target1', 'natlty1_txt',
        'gname', 'weaptype1_txt', 'nkill', 'nwound'
    ]
    df_clean = df[cols_to_keep].copy()

    # Handling missing numerical values by filling with 0
    df_clean['nkill'] = df_clean['nkill'].fillna(0)
    df_clean['nwound'] = df_clean['nwound'].fillna(0)
    df_clean['casualities'] = df_clean['nkill'] + df_clean['nwound']

    print("Data cleaning completed. Cleaned dataframe shape:", df_clean.shape)
    display(df_clean.head())
```

## Comprehensive Visualization Suite
We will now generate multiple visualizations to provide a detailed view of the terrorism landscape.

```python
plt.figure(figsize=(20, 15))

# 1. Top 10 Countries with most attacks
plt.subplot(3, 2, 1)
sns.barplot(y=df_clean['country_txt'].value_counts()[:10].index,
            x=df_clean['country_txt'].value_counts()[:10].values, palette='rocket', hue=df_clean['country_txt'].value_counts()[:10].index, legend=False)
plt.title('Top 10 Affected Countries')

# 2. Attacks by Region
plt.subplot(3, 2, 2)
sns.countplot(y='region_txt', data=df_clean, order=df_clean['region_txt'].value_counts().index, palette='mako', hue='region_txt', legend=False)
plt.title('Attacks by Region')

# 3. Most Common Attack Types
plt.subplot(3, 2, 3)
sns.countplot(y='attacktype1_txt', data=df_clean, order=df_clean['attacktype1_txt'].value_counts().index, palette='magma', hue='attacktype1_txt', legend=False)
plt.title('Most Common Attack Types')

# 4. Top 10 Targeted Sectors
plt.subplot(3, 2, 4)
sns.barplot(y=df_clean['targtype1_txt'].value_counts()[:10].index,
            x=df_clean['targtype1_txt'].value_counts()[:10].values, palette='viridis', hue=df_clean['targtype1_txt'].value_counts()[:10].index, legend=False)
plt.title('Top 10 Targeted Sectors')

# 5. Weapon Types Used
plt.subplot(3, 2, 5)
sns.countplot(y='weaptype1_txt', data=df_clean, order=df_clean['weaptype1_txt'].value_counts().index, palette='flare', hue='weaptype1_txt', legend=False)
plt.title('Weapon Types Used')

# 6. Success vs Failure
plt.subplot(3, 2, 6)
plt.pie(df_clean['success'].value_counts(), labels=['Success', 'Failure'], autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], startangle=90)
plt.title('Proportion of Successful Attacks')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 12))

# 7. Casualty Trends (Killed vs Wounded) Over Years
plt.subplot(2, 1, 1)
casualty_trends = df_clean.groupby('iyear')[['nkill', 'nwound']].sum()
casualty_trends.plot(kind='area', stacked=False, ax=plt.gca(), alpha=0.4)
plt.title('Trend of Fatalities and Injuries (1970-2017)')
plt.ylabel('Number of People')

# 8. Top 10 Most Active Terrorist Groups (excluding Unknown)
plt.subplot(2, 1, 2)
active_groups = df_clean[df_clean['gname'] != 'Unknown']['gname'].value_counts()[:10]
sns.barplot(y=active_groups.index, x=active_groups.values, palette='cubehelix', hue=active_groups.index, legend=False)
plt.title('Top 10 Most Active Terrorist Groups')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 15))

# 9. Monthly Distribution of Attacks (Seasonality check)
plt.subplot(2, 2, 1)
sns.countplot(x='imonth', data=df_clean[df_clean['imonth'] != 0], palette='coolwarm', hue='imonth', legend=False)
plt.title('Attacks by Month (1-12)')

# 10. Suicide Attacks Over Time
plt.subplot(2, 2, 2)
suicide_trends = df_clean.groupby('iyear')['suicide'].sum()
suicide_trends.plot(color='red', marker='o')
plt.title('Trend of Suicide Attacks')

# 11. Regional Trend Shift
plt.subplot(2, 1, 2)
regional_growth = pd.crosstab(df_clean.iyear, df_clean.region_txt)
regional_growth.plot(ax=plt.gca())
plt.title('Regional Attack Trends Over Decades')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

## Advanced Time-Series Analysis of Casualties
This section delves deeper into the temporal patterns of fatalities (`nkill`) and injuries (`nwound`) to identify shifts in the intensity of terrorist activities.

```python
import numpy as np

if 'casualty_trends' in locals() or 'casualty_trends' in globals():
    # 1. Calculating 5-year rolling averages to smooth volatility
    casualty_analysis = casualty_trends.copy()
    casualty_analysis['nkill_rolling'] = casualty_analysis['nkill'].rolling(window=5).mean()
    casualty_analysis['nwound_rolling'] = casualty_analysis['nwound'].rolling(window=5).mean()

    # 2. Calculating Year-over-Year Growth Rate
    casualty_analysis['kill_growth'] = casualty_analysis['nkill'].pct_change() * 100

    plt.figure(figsize=(18, 10))

    # Plotting Rolling Averages
    plt.subplot(2, 1, 1)
    plt.plot(casualty_analysis.index, casualty_analysis['nkill'], label='Annual Fatalities', alpha=0.3, color='red')
    plt.plot(casualty_analysis.index, casualty_analysis['nkill_rolling'], label='5-Year Trend (Fatalities)', linewidth=3, color='red')
    plt.plot(casualty_analysis.index, casualty_analysis['nwound'], label='Annual Injuries', alpha=0.3, color='blue')
    plt.plot(casualty_analysis.index, casualty_analysis['nwound_rolling'], label='5-Year Trend (Injuries)', linewidth=3, color='blue')

    plt.title('Time-Series Trend Analysis: Fatalities vs Injuries (Smoothed)')
    plt.xlabel('Year')
    plt.ylabel('Count')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)

    # Plotting Growth Rates
    plt.subplot(2, 1, 2)
    sns.barplot(x=casualty_analysis.index, y='kill_growth', data=casualty_analysis, palette='RdYlGn_r', hue='kill_growth', legend=False)
    plt.axhline(0, color='black', linewidth=1)
    plt.title('Year-over-Year Percentage Change in Fatalities')
    plt.xlabel('Year')
    plt.ylabel('% Change')
    plt.xticks(rotation=90)

    plt.tight_layout()
    plt.show()

    # Statistical Summary
    print("Casualty Statistics Summary:")
    display(casualty_analysis[['nkill', 'nwound']].describe())
else:
    print("Error: casualty_trends dataframe not found. Please run the previous visualization cells first.")
```

## Expanding the Analysis: Reaching 35 Insight Sets
To achieve the goal of 35 comprehensive insights, we will systematically break down the dataset into specific sub-dimensions including granular target analysis, weapon sub-types, and geographical heatmaps.

```python
plt.figure(figsize=(22, 20))

# 12. Top 15 Specific Target Sub-types
plt.subplot(3, 2, 1)
target_subs = df_clean['targsubtype1_txt'].value_counts()[:15]
sns.barplot(y=target_subs.index, x=target_subs.values, palette='plasma', hue=target_subs.index, legend=False)
plt.title('Insight 12: Top 15 Specific Target Sub-types')

# 13. Distribution of Attacks: Suicide vs Non-Suicide (Pie)
plt.subplot(3, 2, 2)
suicide_counts = df_clean['suicide'].value_counts()
# Dynamic labels to prevent ValueError if only one class exists
labels = ['Non-Suicide', 'Suicide'] if len(suicide_counts) == 2 else ['Non-Suicide']
plt.pie(suicide_counts, labels=labels, autopct='%1.1f%%', colors=['#4CAF50', '#FF5252'][:len(suicide_counts)], explode=[0.1 if i==1 else 0 for i in range(len(suicide_counts))])
plt.title('Insight 13: Prevalence of Suicide Attacks')

# 14. Average Casualties by Attack Type
plt.subplot(3, 2, 3)
avg_cas = df_clean.groupby('attacktype1_txt')['casualities'].mean().sort_values(ascending=False)
sns.barplot(y=avg_cas.index, x=avg_cas.values, palette='cool', hue=avg_cas.index, legend=False)
plt.title('Insight 14: Average Casualties per Attack Type')

# 15. Weapon Sub-category Analysis
plt.subplot(3, 2, 4)
weap_counts = df_clean['weaptype1_txt'].value_counts()[:8]
sns.barplot(y=weap_counts.index, x=weap_counts.values, palette='inferno', hue=weap_counts.index, legend=False)
plt.title('Insight 15: Primary Weapon Categories')

# 16. Fatalities by Region (Boxplot to see outliers)
plt.subplot(3, 2, 5)
sns.boxplot(x='nkill', y='region_txt', data=df_clean[df_clean['nkill'] > 0], palette='Set3', hue='region_txt', legend=False)
plt.xscale('log')
plt.title('Insight 16: Fatality Intensity by Region (Log Scale)')

# 17. Success Rate by Region
plt.subplot(3, 2, 6)
success_region = df_clean.groupby('region_txt')['success'].mean().sort_values()
success_region.plot(kind='barh', color='skyblue')
plt.title('Insight 17: Attack Success Rate by Region')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(22, 12))

# 18. Heatmap of Month vs Region (Attack Frequency)
plt.subplot(1, 2, 1)
region_month = pd.crosstab(df_clean['region_txt'], df_clean['imonth'])
region_month = region_month.drop(0, axis=1) # Drop unknown month
sns.heatmap(region_month, annot=False, cmap='YlOrRd')
plt.title('Insight 18: Monthly Seasonality by Region')

# 19. Cumulative Attacks Over Time
plt.subplot(1, 2, 2)
df_clean.groupby('iyear').size().cumsum().plot(kind='line', color='darkgreen', linewidth=4)
plt.title('Insight 19: Cumulative Growth of Recorded Incidents')
plt.ylabel('Total Attacks')

plt.tight_layout()
plt.show()
```

## Deep Dive: Nationalities, Group Lethality, and Geography (Insights 20-27)
This section expands the analysis to look at the human cost in terms of target nationality and the operational patterns of specific groups.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_theme(style='whitegrid')
plt.figure(figsize=(22, 22))

# 20. Top 15 Victim Nationalities
plt.subplot(3, 2, 1)
natlty_counts = df_clean['natlty1_txt'].value_counts()[:15]
sns.barplot(y=natlty_counts.index, x=natlty_counts.values, hue=natlty_counts.index, palette='viridis', legend=False)
plt.title('Top 15 Victim Nationalities', fontsize=15, fontweight='bold')

# 21. Average Fatalities per Group
plt.subplot(3, 2, 2)
active_g = df_clean[df_clean['gname'] != 'Unknown']['gname'].value_counts()[:10].index
group_lethality = df_clean[df_clean['gname'].isin(active_g)].groupby('gname')['nkill'].mean().sort_values(ascending=False)
sns.barplot(y=group_lethality.index, x=group_lethality.values, hue=group_lethality.index, palette='Reds_r', legend=False)
plt.title('Average Lethality by Top Groups', fontsize=15, fontweight='bold')

# 22. Top 15 Affected Cities
plt.subplot(3, 2, 3)
city_counts = df_clean[df_clean['city'] != 'Unknown']['city'].value_counts()[:15]
sns.barplot(y=city_counts.index, x=city_counts.values, hue=city_counts.index, palette='magma', legend=False)
plt.title('Most Affected Cities', fontsize=15, fontweight='bold')

# 23. Attacks by Day of the Week
plt.subplot(3, 2, 4)
df_temp = df_clean[(df_clean['imonth'] != 0) & (df_clean['iday'] != 0)].copy()
df_temp['date'] = pd.to_datetime(df_temp[['iyear', 'imonth', 'iday']].rename(columns={'iyear': 'year', 'imonth': 'month', 'iday': 'day'}), errors='coerce')
df_temp = df_temp.dropna(subset=['date'])
df_temp['day_name'] = df_temp['date'].dt.day_name()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
sns.countplot(x='day_name', data=df_temp, order=day_order, hue='day_name', palette='coolwarm', legend=False)
plt.title('Attacks by Day of Week', fontsize=15, fontweight='bold')

# 24. Success Rate by Attack Type
plt.subplot(3, 2, 5)
success_attack = df_clean.groupby('attacktype1_txt')['success'].mean().sort_values()
sns.barplot(y=success_attack.index, x=success_attack.values, hue=success_attack.index, palette='RdYlGn', legend=False)
plt.title('Success Rate by Attack Type', fontsize=15, fontweight='bold')

# 25. Geographic Distribution (Fixing Error)
plt.subplot(3, 2, 6)
geo_data = df_clean[['longitude', 'latitude']].dropna()
plt.scatter(geo_data['longitude'], geo_data['latitude'], alpha=0.2, s=2, color='#e74c3c')
plt.title('Geographic Distribution of Attacks', fontsize=15, fontweight='bold')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 10))

# 26. Proportion of Wounded vs Killed per Region
plt.subplot(1, 2, 1)
reg_cas = df_clean.groupby('region_txt')[['nkill', 'nwound']].sum()
reg_cas.plot(kind='barh', stacked=True, ax=plt.gca(), color=['#e74c3c', '#3498db'])
plt.title('Insight 26: Fatalities vs Injuries by Region')

# 27. Targeted Nationalities vs Success
plt.subplot(1, 2, 2)
natlty_success = df_clean[df_clean['natlty1_txt'].isin(natlty_counts.index)].groupby('natlty1_txt')['success'].mean().sort_values()
sns.barplot(y=natlty_success.index, x=natlty_success.values, palette='coolwarm', hue=natlty_success.index, legend=False)
plt.title('Insight 27: Attack Success Rate by Victim Nationality')

plt.tight_layout()
plt.show()
```
